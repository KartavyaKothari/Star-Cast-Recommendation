```python
import pandas as pd
import numpy as np
from collections import Counter
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv('files/castgenre.csv')
```

```python
genre_input = 'fantasy'
a1 = df[df['Genre 1'] == genre_input.capitalize()]
a2 = df[df['Genre 2'] == genre_input.capitalize()]
a3 = df[df['Genre 3'] == genre_input.capitalize()]
c = a1.append(a2, ignore_index=True).append(a3, ignore_index=True)
c.head()
```

```python
actors = c['actors']
x = []
for i in range(len(actors)):
    a = str(actors[i]).split('|')
    a = [z.strip() for z in a]
    x.append(a)
```

```python
te = TransactionEncoder()
te = te.fit(x)
te_ary = te.transform(x)
```

```python
dffd = pd.DataFrame(te_ary,columns=te.columns_)
```

```python
frequent_itemsets = apriori(dffd, min_support =0.0015,use_colnames=True)
frequent_itemsets.to_csv('files/fantasy.csv')
```

```python
rules = association_rules(frequent_itemsets,metric='confidence',min_threshold=0.001)
rules
```

```python
rules[rules['antecedents'].apply(lambda x:set(['Saif Ali Khan']).issubset(x))]
```
